```python
from ultralytics import YOLO

# Tạo model YOLOv8 mới từ đầu (random weights)
model = YOLO('yolov8n.yaml')  

# Huấn luyện
model.train(
    data=r'G:\brain-tumor\brain.yaml',  # đường dẫn đến file yaml
    epochs=100,
    imgsz=512,
    batch=16,
    device=0,  # GPU 0
    name="yolo_brain_tumor_from_scratch",
    project="runs/train",
    optimizer="SGD",
    lr0=0.01,
    degrees=10,
    scale=0.1,
    hsv_h=0.015,
    hsv_s=0.7,
    hsv_v=0.4,
    fliplr=0.5,
    flipud=0.3,
)
```

```python
from ultralytics import YOLO

# Load lại model từ checkpoint
model = YOLO(r'G:/brain-tumor/runs/train/yolo_brain_tumor_from_scratch/weights/last.pt')

# Train tiếp
model.train(
    data=r'G:\brain-tumor\brain.yaml',
    epochs=100,               # thêm 100 epochs nữa
    imgsz=512,
    batch=16,
    device=0,                 # GPU
    name="yolo_brain_tumor_continue",
    project="runs/train",
)
```
